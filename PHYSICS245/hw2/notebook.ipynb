```python
import qutip as qt
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Problem 2
b = qt.Bloch()

for ratio in np.linspace(0, 10, 20):
    # Rewrite formula so that you don't have to divide by (B_x / B_z), in case it's zero
    positive_eigenvector = qt.basis(2, 0) + qt.basis(2, 1) * ratio / (1 + np.sqrt(1 + ratio ** 2))
    b.add_states(positive_eigenvector.unit())

b.show()
```
