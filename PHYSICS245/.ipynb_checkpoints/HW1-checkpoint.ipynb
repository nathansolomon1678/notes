```python
# 1.b
import matplotlib.pyplot as plt
import numpy as np

L = 1
d = 1
v = 1
x = np.linspace(0, L, 100)

def y(x, t, A):
    sum = x*0
    for n in range(len(A)):
        sum += A[n] * np.sin(x * n * np.pi / L) * np.cos(n * np.pi * v * t / L)
    return sum

def get_A(n):
    # gives the nth coefficient
    if n % 2 == 0:
        return 0
    val = 8 * d / n**2 / np.pi**2
    if n % 4 in [0, 3]:
        val *= -1
    return val

A = [get_A(n) for n in range(20)]
print(A)

def plot_y(x, times, A):
    for t in times:
        plt.plot(y(x, t, A))
    plt.show()

plot_y(x, [0, .1, .2, .3, .4, .5, .6, .7], A)
plot_y(x, [0, .1, .2, .3, .4, .5, .6, .7], A[:6])
plt.show()
```

```python
# QUantum Toolbox In Python
import qutip as qt

def print_qobj(qutip_object):
    # Prints just the matrix data, without the other info
    assert isinstance(qt.sigmax(), qt.core.qobj.Qobj)
    print(qutip_object[:, :])
```

```python
# 6.a
# For convenience, assume hbar=1 throughout this course, so
# S = sigma / 2
S_x = qt.sigmax() / 2
S_y = qt.sigmay() / 2
S_z = qt.sigmaz() / 2
print_qobj(S_x)
print('\n')
print_qobj(S_y)
print('\n')
print_qobj(S_z)
```

```python
# 6.b
Psi_1 = qt.basis(2, 0)
print_qobj(Psi_1)
print('\n')
print(qt.expect(qt.sigmaz() / 2, Psi_1))
```

```python
# 6.c
Psi_2 = (qt.basis(2, 0) + qt.basis(2, 1)).unit()
print_qobj(Psi_2)
print('\n')
print(qt.expect(qt.sigmaz() / 2, Psi_2))
```

```python
# 6.d
def uncertainty(operator, state):
    return np.sqrt(qt.expect(operator * operator, state) - qt.expect(operator, state) ** 2)

print(uncertainty(S_z, Psi_1))
```

```python
# 6.e
print(uncertainty(S_z, Psi_2))
```

```python
# 6.f
b = qt.Bloch()
b.add_states(Psi_1) # plot in green
b.add_states(Psi_2) # plot in orange
b.show()
print("\n|+X> is Psi_2 because it is the eigenvector of S_x with eigenvalue +1.")
```
